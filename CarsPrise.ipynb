```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(outliers)
```

```python
# Definir las columnas relevantes
columnas_relevantes = ["price", "enginesize", "horsepower", "citympg", "highwaympg"]
# Convertir las columnas a tipo numérico
df[columnas_relevantes] = df[columnas_relevantes].astype(float)

# Buscar outliers en las columnas relevantes
for columna in columnas_relevantes:
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]
    print(f"Outliers en la columna {columna}:")
    print(outliers)
    print("\n")
# Este código recorre cada una de las columnas relevantes y calcula el rango intercuartil (IQR). Luego, define los límites inferior y superior del rango aceptable multiplicando el IQR por 1.5. Finalmente, identifica los valores que están por debajo del límite inferior o por encima del límite superior como valores atípicos y los imprime en la salida.

# Recuerda reemplazar "ruta_del_archivo.csv" con la ruta y el nombre de tu archivo CSV que contiene los datos. Además, asegúrate de tener pandas instalado en tu entorno de Python para poder ejecutar el código correctamente.






```
